```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Organize Data with Python
# Movie data: A list of dictionaries with movie attributes
movies = [
    {'Movie Title': 'The Shawshank Redemption', 'Genre': 'Drama', 'Rating': 9.3, 'Release Year': 1994},
    {'Movie Title': 'The Dark Knight', 'Genre': 'Action', 'Rating': 9.0, 'Release Year': 2008},
    {'Movie Title': 'Forrest Gump', 'Genre': 'Drama', 'Rating': 8.8, 'Release Year': 1994},
    {'Movie Title': 'The Godfather', 'Genre': 'Crime', 'Rating': 9.2, 'Release Year': 1972},
    {'Movie Title': 'Pulp Fiction', 'Genre': 'Crime', 'Rating': 8.9, 'Release Year': 1994},
    {'Movie Title': 'The Matrix', 'Genre': 'Sci-Fi', 'Rating': 8.7, 'Release Year': 1999},
    {'Movie Title': 'Inception', 'Genre': 'Sci-Fi', 'Rating': 8.8, 'Release Year': 2010},
    {'Movie Title': 'The Lion King', 'Genre': 'Animation', 'Rating': 8.5, 'Release Year': 1994},
]

# User data: A dictionary where each user has a list of preferred genres
user_profiles = {
    'User1': ['Action', 'Adventure', 'Sci-Fi'],
    'User2': ['Romance', 'Drama', 'Comedy'],
    'User3': ['Horror', 'Thriller'],
}

# Step 2: Convert movie list into DataFrame for easy manipulation with Pandas
movies_df = pd.DataFrame(movies)

# Step 3: Data Cleaning with Pandas
# Check for missing values
print("\nChecking for missing values:")
print(movies_df.isnull().sum())

# Drop rows with missing genre or movie title
movies_df = movies_df.dropna(subset=['Genre', 'Movie Title'])

# Remove duplicates (if any)
movies_df = movies_df.drop_duplicates(subset=['Movie Title'])

# Standardize genre names (example: 'Sci-Fi' -> 'Science Fiction')
movies_df['Genre'] = movies_df['Genre'].replace({'Sci-Fi': 'Science Fiction'})

# Preview the cleaned dataset
print("\nCleaned dataset:")
print(movies_df)

# Step 4: Visualizing Genre and Ratings Trends
# Genre Frequency: Bar Chart to show most common genres
plt.figure(figsize=(8, 6))
genre_counts = movies_df['Genre'].value_counts()
sns.barplot(x=genre_counts.index, y=genre_counts.values, palette='viridis')
plt.title('Genre Frequency')
plt.xlabel('Genre')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()

# Rating Distribution: Histogram to show ratings distribution
plt.figure(figsize=(8, 6))
sns.histplot(movies_df['Rating'], bins=10, kde=True, color='blue')
plt.title('Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()

# Step 5: Movie Recommendations based on User Preferences
# For each user, recommend movies that match their preferred genres
for user, preferred_genres in user_profiles.items():
    print(f"\nRecommended movies for {user}:")
    recommended_movies = movies_df[movies_df['Genre'].apply(lambda genre: genre in preferred_genres)]
    print(recommended_movies[['Movie Title', 'Genre', 'Rating']])

# Step 6: Saving the Cleaned Data
movies_df.to_csv('cleaned_movies_dataset.csv', index=False)
print("\nCleaned dataset saved as 'cleaned_movies_dataset.csv'.")
```

```python
import sqlite3
import pandas as pd

# Create a new SQLite database or connect to an existing one
conn = sqlite3.connect('movies_recommendation.db')
cursor = conn.cursor()

# Create tables for movies and user preferences
cursor.execute('''
CREATE TABLE IF NOT EXISTS movies (
    id INTEGER PRIMARY KEY,
    title TEXT,
    genre TEXT,
    rating REAL,
    release_year INTEGER
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS user_preferences (
    user_id INTEGER PRIMARY KEY,
    preferred_genres TEXT
)
''')

# Insert movie data into the 'movies' table
movies = [
    ('The Shawshank Redemption', 'Drama', 9.3, 1994),
    ('The Dark Knight', 'Action', 9.0, 2008),
    ('Forrest Gump', 'Drama', 8.8, 1994),
    ('The Godfather', 'Crime', 9.2, 1972),
    ('Pulp Fiction', 'Crime', 8.9, 1994),
    ('The Matrix', 'Sci-Fi', 8.7, 1999),
    ('Inception', 'Sci-Fi', 8.8, 2010),
    ('The Lion King', 'Animation', 8.5, 1994)
]

cursor.executemany('''
INSERT INTO movies (title, genre, rating, release_year)
VALUES (?, ?, ?, ?)
''', movies)

# Insert user preferences data into the 'user_preferences' table
user_profiles = {
    1: ['Action', 'Adventure', 'Sci-Fi'],   # Using integer for user_id
    2: ['Romance', 'Drama', 'Comedy'],
    3: ['Horror', 'Thriller'],
}

for user_id, genres in user_profiles.items():
    cursor.execute('''
    INSERT INTO user_preferences (user_id, preferred_genres)
    VALUES (?, ?)
    ''', (user_id, ','.join(genres)))

# Commit changes and close the connection
conn.commit()

# Step 2: SQL Query to Find Matching Movies Based on User Preferences
def get_movie_recommendations(user_id):
    cursor.execute('''
    SELECT m.title, m.genre, m.rating
    FROM movies m
    JOIN user_preferences u ON ',' || u.preferred_genres || ',' LIKE '%,' || m.genre || ',%'
    WHERE u.user_id = ?
    ''', (user_id,))
    
    return cursor.fetchall()

# Example: Get movie recommendations for User1
user_id = 1
recommended_movies = get_movie_recommendations(user_id)

print(f"\nMovies recommended for User{user_id}:")
for movie in recommended_movies:
    print(f"Title: {movie[0]}, Genre: {movie[1]}, Rating: {movie[2]}")

# Close the connection
conn.close()
```
